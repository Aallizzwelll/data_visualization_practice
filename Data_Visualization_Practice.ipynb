```python
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_performance_df = pd.read_csv("C:/Users/nh010/LOCAL/Data Analysis/student_performance_from_Kaggle.zip", compression = 'zip')
print(student_performance_df.head())
```

```python
#gender distribution pie charts
#pandas pie chart
gender_counts = student_performance_df['gender'].value_counts()
gender_counts.plot(kind ='pie', title = 'Gender of students',  autopct='%1.1f%%', legend = False, ylabel="")

#matplotlib pie chart
labels = gender_counts.index
sizes = gender_counts.values
colors = ['#bcc2cf','#3181ff']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=360)
plt.title('Gender Distribution of Students')
plt.show()


```

```python
#ethnicity pie chart
ethnicity_counts = student_performance_df['race/ethnicity'].value_counts()
ethnicity_counts.plot(kind ='pie', title = 'ethnicity of students', autopct='%1.1f%%', legend = False, ylabel="")
plt.show()
```

```python
#ethnicity bar chart
ethnicity_counts = student_performance_df['race/ethnicity'].value_counts().sort_index()
ethnicity_counts.plot.bar(x='ethnicity_counts', color='green')
plt.title('Race/Ethnicity of Students')
plt.show()
```

```python
testprep_counts = student_performance_df['test preparation course'].value_counts()
testprep_counts.plot(kind ='pie', title = 'test preparation status',  autopct='%1.1f%%', legend = False, ylabel="")
plt.show()
```

```python
lunch_status_counts = student_performance_df['lunch'].value_counts()
lunch_status_counts.plot(kind ='pie', title = 'free/reduced lunch status',  autopct='%1.1f%%', legend = False, ylabel="")
plt.show()
```

```python
parent_edu_counts = student_performance_df['parental level of education'].value_counts()
parent_edu_counts.plot(kind ='bar', title = 'parental level of education', legend = False, ylabel="")
plt.show()
```

```python
new_ordered_edu = ['some high school', 'high school', "associate's degree", 'some college', "bachelor's degree", "master's degree"]
parent_edu_counts = student_performance_df['parental level of education'].value_counts().reindex(new_ordered_edu)
parent_edu_counts.plot(kind ='bar', title = 'parental level of education', legend = False, ylabel="",xlabel="")
plt.show()
```

```python
#Average MATH by ethnicity
grouped_ethnicity = student_performance_df.groupby('race/ethnicity').agg({'math score':'mean'})
grouped_ethnicity = grouped_ethnicity.reset_index()
grouped_ethnicity =grouped_ethnicity.sort_values(by='race/ethnicity', ascending=True)
grouped_ethnicity.plot(kind='bar',x='race/ethnicity',y='math score', color='blue', title = 'Average math performance by ethnicty grouping', legend=False,xlabel="")

#Average Reading by ethnicity
grouped_ethnicity = student_performance_df.groupby('race/ethnicity').agg({'reading score':'mean'})
grouped_ethnicity = grouped_ethnicity.reset_index()
grouped_ethnicity =grouped_ethnicity.sort_values(by='race/ethnicity', ascending=True)
grouped_ethnicity.plot(kind='bar',x='race/ethnicity',y='reading score', color='green', title = 'Average reading performance by ethnicty grouping', legend=False,xlabel="")

#Average Writing by ethnicity
grouped_ethnicity = student_performance_df.groupby('race/ethnicity').agg({'writing score':'mean'})
grouped_ethnicity = grouped_ethnicity.reset_index()
grouped_ethnicity =grouped_ethnicity.sort_values(by='race/ethnicity', ascending=True)
grouped_ethnicity.plot(kind='bar',x='race/ethnicity',y='writing score', color='red', title = 'Average writing performance by ethnicty grouping', legend=False,xlabel="")
plt.show()
```

```python

#average math score by gender
sns.barplot(x='gender', y='math score', data=student_performance_df, hue= 'gender', legend=False, palette='dark')
plt.title('Average math score by gender')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#average reading score by gender
sns.barplot(x='gender', y='reading score', data=student_performance_df, legend=False, hue = 'gender', palette='Set2')
plt.title('average reading score by gender')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#average writing score by gender
sns.barplot(x='gender', y='writing score', data=student_performance_df, legend=False, hue = 'gender', palette='Set3')
plt.title('average writing score by gender')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
